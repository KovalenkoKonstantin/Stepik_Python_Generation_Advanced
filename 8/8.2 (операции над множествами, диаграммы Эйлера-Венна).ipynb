```python
from matplotlib import pyplot as plt
import numpy as np
from matplotlib_venn import venn3, venn3_circles

plt.figure(figsize=(4, 4))
v = venn3(subsets=(1, 1, 1, 1, 1, 1, 1), set_labels=('A', 'B', 'C'))
v.get_patch_by_id('100').set_alpha(1.0)
v.get_patch_by_id('100').set_color('white')
v.get_label_by_id('100').set_text('Unknown')
v.get_label_by_id('A').set_text('Set "A"')
c = venn3_circles(subsets=(1, 1, 1, 1, 1, 1, 1), linestyle='dashed')
c[0].set_lw(1.0)
c[0].set_ls('dotted')
plt.title("Sample Venn diagram")
plt.annotate('Unknown set', xy=v.get_label_by_id('100').get_position() - np.array([0, 0.05]), xytext=(-70, -70),
             ha='center', textcoords='offset points', bbox=dict(boxstyle='round,pad=0.5', fc='gray', alpha=0.1),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5', color='gray'))
plt.show()
```

```python
import sympy as sp

# Определяем переменные
x, i, j, k = sp.symbols('EnglishDeutschFrancais EnglishDeutsch  EnglishFrancais DeutschFrancais')

# Задаем уравнения
eq1 = 30 - i - j - k - (25 - 1)
eq2 = i + x - 6
eq3 = j + x - 10
eq4 = k + x - 4

# Решаем систему уравнений
solution = sp.solve((eq1, eq2, eq3, eq4), (x, i, j, k))

# Выводим решение
print(solution)
```

```python
n, m, k, x, y, z, t, a = (int(input()) for _ in range(8))

only_1_and_2 = n + m - x - t  # те, кто прочитали только первую и вторую книги
only_2_and_3 = m + k - y - t  # те, кто прочитали только вторую и третью книги
only_1_and_3 = n + k - z - t  # те, кто прочитали только первую и третью книги

only_two = only_1_and_2 + only_2_and_3 + only_1_and_3  # те, кто прочитали только 2 книги

only_1 = n - only_1_and_2 - only_1_and_3 - t  # те, кто прочитал только первую книгу
only_2 = m - only_1_and_2 - only_2_and_3 - t  # те, кто прочитал только вторую книгу
only_3 = k - only_1_and_3 - only_2_and_3 - t  # те, кто прочитал только третью книгу

only_one = only_1 + only_2 + only_3  # те, кто прочитал только 1 книгу

read_nothing = a - only_one - only_two - t  # те, кто ничего не прочитали

print(only_one)
print(only_two)
print(read_nothing)
```

```python
from collections import Counter

# –í–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö
data = input().strip()

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏ –≤ —Å–ø–∏—Å–æ–∫ —á–∏—Å–µ–ª
numbers = list(map(int, data.split()))

# –ü–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—Ö–æ–∂–¥–µ–Ω–∏–π –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–∞
count = Counter(numbers)

# –°—É–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π
repeated_count = sum(value - 1 for value in count.values() if value > 1)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print(repeated_count)
```

```python
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–Ω–∏–≥ –≤ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–µ
m = int(input().strip())
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–Ω–∏–≥ –≤ —Å–ø–∏—Å–∫–µ –Ω–∞ –ª–µ—Ç–æ
n = int(input().strip())

# –°–æ–∑–¥–∞–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π –∫–Ω–∏–≥ –∏–∑ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
library_books = set()

# –°—á–∏—Ç—ã–≤–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –∫–Ω–∏–≥ –∏–∑ –¥–æ–º–∞—à–Ω–µ–π –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
for _ in range(m):
    book = input().strip()
    library_books.add(book)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –∫–Ω–∏–≥–∏ –∏–∑ –ª–µ—Ç–Ω–µ–≥–æ —Å–ø–∏—Å–∫–∞ –∏ –≤—ã–¥–∞–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
for _ in range(n):
    summer_book = input().strip()
    if summer_book in library_books:
        print("YES")
    else:
        print("NO")
```

```python
# –°—á–∏—Ç—ã–≤–∞–µ–º —Å—Ç—Ä–æ–∫–∏ —Å —á–∏—Å–ª–∞–º–∏
first_list = input().strip().split()
second_list = input().strip().split()

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç—Ä–æ–∫–∏ –≤ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª
first_set = set(map(int, first_list))
second_set = set(map(int, second_list))

# –ù–∞—Ö–æ–¥–∏–º –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ –¥–≤—É—Ö –º–Ω–æ–∂–µ—Å—Ç–≤
common_numbers = first_set.intersection(second_set)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –æ–±—â–∏–µ —á–∏—Å–ª–∞ –∏ –≤—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
if common_numbers:
    # –°–æ—Ä—Ç–∏—Ä—É–µ–º —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è –∏ –≤—ã–≤–æ–¥–∏–º
    sorted_numbers = sorted(common_numbers, reverse=True)
    print(" ".join(map(str, sorted_numbers)))
else:
    print("BAD DAY")
```

```python
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –º–∞—Ç–µ–º–∞—Ç–∏–∫—É
m = int(input().strip())
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É
n = int(input().strip())

# –°–æ–∑–¥–∞–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –¥–ª—è —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –º–∞—Ç–µ–º–∞—Ç–∏–∫—É –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É
math_students = set()
info_students = set()

# –°—á–∏—Ç—ã–≤–∞–µ–º —Ñ–∞–º–∏–ª–∏–∏ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –º–∞—Ç–µ–º–∞—Ç–∏–∫—É
for _ in range(m):
    student_name = input().strip()
    math_students.add(student_name)

# –°—á–∏—Ç—ã–≤–∞–µ–º —Ñ–∞–º–∏–ª–∏–∏ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É
for _ in range(n):
    student_name = input().strip()
    info_students.add(student_name)

# –ù–∞—Ö–æ–¥–∏–º —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö —Ç–æ–ª—å–∫–æ –º–∞—Ç–µ–º–∞—Ç–∏–∫—É
only_math_students = math_students - info_students
# –ù–∞—Ö–æ–¥–∏–º —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö —Ç–æ–ª—å–∫–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É
only_info_students = info_students - math_students

# –°—á–∏—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç
only_one_subject_students = len(only_math_students) + len(only_info_students)

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
if only_one_subject_students > 0:
    print(only_one_subject_students)
else:
    print("NO")
```

```python
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –º–∞—Ç–µ–º–∞—Ç–∏–∫—É
m = int(input())
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–∫—É
n = int(input())

students = {input() for i in range(n + m)}

# –°—á–∏—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—á–µ–Ω–∏–∫–æ–≤, –∏–∑—É—á–∞—é—â–∏—Ö —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –ø—Ä–µ–¥–º–µ—Ç
only_one_subject_students = n + m - (n + m - len(students)) * 2

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
if only_one_subject_students > 0:
    print(only_one_subject_students)
else:
    print("NO")
```

```python
# –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É—Ä–æ–∫–æ–≤
m = int(input())

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Å —Ñ–∞–º–∏–ª–∏—è–º–∏ —É—á–µ–Ω–∏–∫–æ–≤, –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞–≤—à–∏—Ö –Ω–∞ –≤—Å–µ—Ö —É—Ä–æ–∫–∞—Ö
present_students = set()

# –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –∫–∞–∂–¥—ã–π —É—Ä–æ–∫
for i in range(m):
    # –°—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ–∞–º–∏–ª–∏–π –Ω–∞ —Ç–µ–∫—É—â–µ–º —É—Ä–æ–∫–µ
    n_i = int(input())
    
    # –°—á–∏—Ç—ã–≤–∞–µ–º —Ñ–∞–º–∏–ª–∏–∏ —É—á–µ–Ω–∏–∫–æ–≤ –Ω–∞ —Ç–µ–∫—É—â–µ–º —É—Ä–æ–∫–µ
    current_students = set()
    for _ in range(n_i):
        current_students.add(input().strip())
    
    # –ï—Å–ª–∏ —ç—Ç–æ –ø–µ—Ä–≤—ã–π —É—Ä–æ–∫, –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—â–∏—Ö
    if i == 0:
        present_students = current_students
    else:
        # –ü–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—â–∏—Ö
        present_students.intersection_update(current_students)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫
result = sorted(present_students)

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
for student in result:
    print(student)
```

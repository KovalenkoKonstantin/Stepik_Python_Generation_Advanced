```python
from collections import Counter

# Ввод данных
data = input().strip()

# Преобразование строки в список чисел
numbers = list(map(int, data.split()))

# Подсчет количества вхождений каждого числа
count = Counter(numbers)

# Суммирование количества повторений
repeated_count = sum(value - 1 for value in count.values() if value > 1)

# Вывод результата
print(repeated_count)
```

```python
n = int(input().strip())

# Создаем множество для хранения уникальных названных городов
cities = set()

# Считываем названные города и добавляем их в множество
for _ in range(n):
    city = input().strip()
    cities.add(city)

# Считываем новый город
new_city = input().strip()

# Проверяем, был ли этот город уже назван
if new_city in cities:
    print("REPEAT")
else:
    print("OK")
    cities.add(new_city)  # добавляем новый город в множество
```

```python
# Считываем количество книг в домашней библиотеке
m = int(input().strip())
# Считываем количество книг в списке на лето
n = int(input().strip())

# Создаем множество для хранения названий книг из домашней библиотеки
library_books = set()

# Считываем названия книг из домашней библиотеки
for _ in range(m):
    book = input().strip()
    library_books.add(book)

# Проверяем книги из летнего списка и выдаем результат
for _ in range(n):
    summer_book = input().strip()
    if summer_book in library_books:
        print("YES")
    else:
        print("NO")
```

```python
# Считываем строки с числами
first_list = input().strip().split()
second_list = input().strip().split()

# Преобразуем строки в множества целых чисел
first_set = set(map(int, first_list))
second_set = set(map(int, second_list))

# Находим пересечение двух множеств
common_numbers = first_set.intersection(second_set)

# Проверяем, есть ли общие числа и выводим результат
if common_numbers:
    # Сортируем числа в порядке убывания и выводим
    sorted_numbers = sorted(common_numbers, reverse=True)
    print(" ".join(map(str, sorted_numbers)))
else:
    print("BAD DAY")
```

```python
# Считываем количество учеников, изучающих математику
m = int(input().strip())
# Считываем количество учеников, изучающих информатику
n = int(input().strip())

# Создаем множества для учеников, изучающих математику и информатику
math_students = set()
info_students = set()

# Считываем фамилии учеников, изучающих математику
for _ in range(m):
    student_name = input().strip()
    math_students.add(student_name)

# Считываем фамилии учеников, изучающих информатику
for _ in range(n):
    student_name = input().strip()
    info_students.add(student_name)

# Находим учеников, изучающих только математику
only_math_students = math_students - info_students
# Находим учеников, изучающих только информатику
only_info_students = info_students - math_students

# Считаем общее количество учеников, изучающих только один предмет
only_one_subject_students = len(only_math_students) + len(only_info_students)

# Выводим результат
if only_one_subject_students > 0:
    print(only_one_subject_students)
else:
    print("NO")
```

```python
# Считываем фамилии, записанные руководителем и помощником
director_names = input().strip().split()
assistant_names = input().strip().split()

# Объединяем оба списка в одно множество для устранения дубликатов
unique_names = set(director_names).union(set(assistant_names))

# Преобразуем множество обратно в список и сортируем его
sorted_names = sorted(unique_names)

# Выводим фамилии в лексикографическом порядке
print(" ".join(sorted_names))
```

```python
# Считываем количество уроков
m = int(input())

# Инициализируем множество с фамилиями учеников, присутствовавших на всех уроках
present_students = set()

# Обрабатываем каждый урок
for i in range(m):
    # Считываем количество фамилий на текущем уроке
    n_i = int(input())
    
    # Считываем фамилии учеников на текущем уроке
    current_students = set()
    for _ in range(n_i):
        current_students.add(input().strip())
    
    # Если это первый урок, инициализируем множество присутствующих
    if i == 0:
        present_students = current_students
    else:
        # Пересчитываем множество присутствующих
        present_students.intersection_update(current_students)

# Преобразуем множество в отсортированный список
result = sorted(present_students)

# Выводим результат
for student in result:
    print(student)
```

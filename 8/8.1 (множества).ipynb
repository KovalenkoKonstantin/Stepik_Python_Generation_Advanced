```python
from math import sqrt

x, y = 36, 4
print(sqrt(x * y))
print(sqrt(x / y))
print(sqrt(x + y))
print(sqrt(x - y))
```

```python
def what_kind_of_catet(c: int, a: int) -> int:
    return (c ** 2 - a ** 2) ** 0.5


print(what_kind_of_catet(4, 3))  # A = 2.6457513110645907
print(what_kind_of_catet(5, 4))  # B = 3.0
print(what_kind_of_catet(6, 5))  # C = 3.3166247903554
print(what_kind_of_catet(7, 6))  # D = 3.605551275463989
```
